```python
import requests 
from bs4 import BeautifulSoup as bsb
import csv
import time

from selenium import webdriver
```

```python
class IMDBReviewExtraction:
    
    def __init__(self,url):
        time.sleep(5)
        data = requests.get(url).text
        self.html = bsb(data,'html.parser')
    
    def ReviewExtraction(self):
        m=1
        reviews = self.html.find_all('div',{'class':'lister'})
        
        for review in reviews:
            tt = review.find_all('div',{'class':'text show-more__control'})
            for t in tt:
                print(t.text)
                print(m)
                print()
                m+=1
        
```

```python
url = 'https://www.imdb.com/title/tt0499549/reviews?ref_=tt_urv'
```

```python
imbd = IMDBReviewExtraction(url)
```

```python
imbd.ReviewExtraction()
```

```python
webdriver = webdriver.Firefox()
webdriver.get(url)
html = webdriver.page_source.encode('utf-8')
num = 0
while num<=50:
    webdriver.find_element_by_class_name('ipl-load-more__button').click()
    num+=1
    time.sleep(3)
html = webdriver.page_source.encode('utf-8')
webdriver.close()
```

```python
html = bsb(html,'html.parser')
```

```python
 m=1
reviews = html.find_all('div',{'class':'lister'})
with open('IMDBReviewData.csv','a')as file:
    csvFile = csv.writer(file)
    for review in reviews:
        tt = review.find_all('div',{'class':'text show-more__control'})
        for t in tt:
            csvFile.writerow([t.text])
            print(m)
            print()
            m+=1
```
